```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/colorful-paper-flower-abstract-fir-background.jfif")
```

```python
plt.figure(figsize=(20,8))
plt.imshow(img)
plt.show()
```


Opencv loads the image and colour format in bgr and not in the rgb format

```python
#converting the form to rgb format
RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(RGB)
plt.show()
```

```python
#converting the form to HSV format
HSV = cv2.cvtColor(RGB, cv2.COLOR_RGB2HSV)
```

```python
#predicting yellow colour
lower = np.array([25, 150, 50])
upper = np.array([35, 255, 255])
```

```python
mask = cv2.inRange(HSV, lower, upper)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(mask)
plt.show()
```

```python

res = cv2.bitwise_and(RGB, RGB, mask=mask)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(res)
plt.show()
```

```python
img2 = cv2.imread("/content/green-nature-pictures.jpeg")
plt.figure(figsize=(20,8))
plt.imshow(img2)
plt.show()
```

```python
#converting the form to rgb format
RGB = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,8))
plt.imshow(RGB)
plt.show()
```

```python
#converting the form to HSV format
HSV = cv2.cvtColor(RGB, cv2.COLOR_RGB2HSV)
```

```python
#predicting green colour
lower = np.array([35, 150, 50])
upper = np.array([75, 255, 255])
```

```python
mask = cv2.inRange(HSV, lower, upper)
plt.figure(figsize=(20,8))
plt.imshow(mask)
plt.show()
```

```python
res = cv2.bitwise_and(RGB, RGB, mask=mask)
plt.figure(figsize=(20,8))
plt.imshow(res)
plt.show()
```
